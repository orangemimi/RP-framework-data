```python
import numpy as np
import matplotlib.pyplot as plt

# 假设我们有过去一年的日降水量数据（单位：毫米）
# 这里生成一些模拟数据，假设均值为5毫米，标准差为2毫米
np.random.seed(42)
historical_rainfall = np.random.normal(5, 2, 365)

# 计算历史数据的均值和标准差
mu = np.mean(historical_rainfall)
sigma = np.std(historical_rainfall)

# 设置蒙特卡罗模拟的参数
num_simulations = 10000  # 模拟次数
num_days = 7  # 预测未来7天

# 生成未来7天的降水量样本
simulated_rainfall = np.random.normal(mu, sigma, (num_simulations, num_days))

# 计算每一天的预测均值和置信区间
mean_rainfall = np.mean(simulated_rainfall, axis=0)
lower_bound = np.percentile(simulated_rainfall, 2.5, axis=0)
upper_bound = np.percentile(simulated_rainfall, 97.5, axis=0)

# 可视化结果
days = np.arange(1, num_days + 1)
plt.figure(figsize=(10, 6))
plt.plot(days, mean_rainfall, color='blue', label='平均降水量')
plt.fill_between(days, lower_bound, upper_bound, color='lightblue', alpha=0.5, label='95%置信区间')
plt.xlabel('未来天数')
plt.ylabel('降水量 (毫米)')
plt.title('未来7天降水量预测（蒙特卡罗模拟）')
plt.legend()
plt.grid(True)
plt.show()
```
