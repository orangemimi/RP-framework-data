```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings 

warnings.filterwarnings("ignore")

# 生成等间距的x值数组，范围从0到2π，包含300个点
x = np.linspace(0, 2 * np.pi, 300)

# 生成Out0列数据，基本符合sin(x)分布并加上随机扰动
out0 = np.cos(x) + 2* np.random.normal(0, 0.1, len(x))

# 初始化一个数据框
data = pd.DataFrame()

# 生成Out1到Out1000列数据，基本符合cos(x)分布并加上随机扰动
for i in range(1, 1001):
    data[f'Out{i}'] = np.cos(x) + np.random.normal(0, 0.1, len(x))

# 查看数据框
# print(data)

# 绘制图片
plt.figure(figsize=(50, 10))

# 绘制Out0数据
plt.plot(x, out0, label='Out0 (sin(x) with noise)', color='blue')

# 绘制Out1到Out5数据作为示例
for i in range(1, 6):
    plt.plot(x, data[f'Out{i}'], label=f'Out{i} (cos(x) with noise)', linestyle='--')

plt.legend()
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Generated Data: Out0 (sin(x) with noise) and Out1-1000 (cos(x) with noise)')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings 

warnings.filterwarnings("ignore")

# 生成Out1到Out1000列数据，基本符合cos(x)分布并加上随机扰动
for i in range(1, 1001):
    data[f'Out{i}'] = np.cos(x) + np.random.normal(0, 0.1, len(x))

# 计算每一行的均值和标准差
means = data.mean(axis=1)
stds = data.std(axis=1)

# 计算缓冲区的上下界（均值 ± 3 * 标准差）
buffer_upper = means + 3 * stds
buffer_lower = means - 3 * stds

# 绘制图片
plt.figure(figsize=(50, 10))

# 绘制缓冲区范围
plt.fill_between(np.arange(len(means)), buffer_lower, buffer_upper, color='lightgray', alpha=0.5, label='Buffer Zone')

# 绘制Out1到Out5数据作为示例
for i in range(1, 6):
    plt.plot(np.arange(len(means)), data[f'Out{i}'], label=f'Out{i}', linestyle='--')

plt.legend()
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Generated Data and Buffer Zone')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# 初始化缓冲区上下界的数组
buffer_upper = []
buffer_lower = []

# 对每一行进行正态性检验并计算缓冲区
for i in range(data.shape[0]):
    row_data = data.iloc[i, :]
    
    # Shapiro-Wilk检验
    stat, p_value = shapiro(row_data)
    
    if p_value > 0.05:
        # 数据符合正态分布，使用±2个标准差
        print("yes")
        mean = row_data.mean()
        std = row_data.std()
        upper = mean + 2 * std
        lower = mean - 2 * std
    else:
        print("no")
        # 数据不符合正态分布，使用IQR
        q1 = row_data.quantile(0.25)
        q3 = row_data.quantile(0.75)
        iqr = q3 - q1
        upper = q3 + 1.5 * iqr
        lower = q1 - 1.5 * iqr
    
    buffer_upper.append(upper)
    buffer_lower.append(lower)

# 将缓冲区上下界转换为数组
buffer_upper = np.array(buffer_upper)
buffer_lower = np.array(buffer_lower)

# 生成x轴
x = np.arange(data.shape[0])

# 绘制缓冲区范围
plt.figure(figsize=(50, 10))
# 绘制Out0数据
plt.plot(x, out0, label='Out0 (sin(x) with noise)', color='blue')

plt.fill_between(x, buffer_lower, buffer_upper, color='lightgray', alpha=0.5, label='Buffer Zone')

# 绘制部分数据列以示例
for i in range(5):
    plt.plot(data.index, data.iloc[:, i], label=f'Out{i}', linestyle='--')

plt.legend()
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Buffer Zone with Mixed Methods (Normal and IQR)')
plt.show()
```

```python
# 计算Out0在缓冲区内的比例
within_buffer = np.logical_and(out0 >= buffer_lower, out0 <= buffer_upper)
probability_within_buffer = np.mean(within_buffer)
print(within_buffer)
print(probability_within_buffer)
```

```python


# 生成所需列的列名列表
# column_names = ['Out0'] + [f'Out{i}' for i in range(1, 1001)]

# 筛选所需的列
# data_df = df[column_names]
data_df=data
data_df['Out0']=out0

# print(data_df)

def lins_ccc(x, y):
    """计算Lin's Concordance Correlation Coefficient (CCC)"""
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    var_x = np.var(x)
    var_y = np.var(y)
    cov_xy = np.mean((x - mean_x) * (y - mean_y))
    rho = cov_xy / np.sqrt(var_x * var_y)
    ccc = (2 * rho * np.sqrt(var_x) * np.sqrt(var_y)) / (var_x + var_y + (mean_x - mean_y)**2)
    return ccc
    

# 计算R0, R1, ..., R1000
R_values = {}
CCC_values={}
for i in range(1, 1001):
    R_values[f'R{i}'] = r2_score(data_df['Out0'], data_df[f'Out{i}'])
    CCC_values[f'R{i}'] = lins_ccc(data_df['Out0'], data_df[f'Out{i}'])

# print(R_values)
# bootstrap抽样并计算R²
bootstrap_R_values = []
bootstrap_CCC_values=[]
num_bootstrap_samples = 10000

for _ in range(num_bootstrap_samples):
    sample_indices = np.random.choice(range(1, 1001), size=2, replace=True)
    # print(sample_indices)
    sample1 = data_df.iloc[:, sample_indices[0]]
    sample2 = data_df.iloc[:, sample_indices[1]]
    # 使用训练集和验证集计算Lin's CCC
    # train_sample1 = data_df.iloc[train_indices, 1]
    # train_sample2 = data_df.iloc[train_indices, 2]
    # test_sample1 = data_df.iloc[test_indices, 1]
    # test_sample2 = data_df.iloc[test_indices, 2]
    
    bootstrap_R_values.append(r2_score(sample1, sample2))
    bootstrap_CCC_values.append(lins_ccc(sample1, sample2))

# 蒙特卡洛随机检验
alpha = 0.05
bootstrap_R_values = np.sort(bootstrap_R_values)
R_threshold_low = np.percentile(bootstrap_R_values, alpha/2*100)
R_threshold_high = np.percentile(bootstrap_R_values, (1-alpha/2)*100)

bootstrap_CCC_values = np.sort(bootstrap_CCC_values)
CCC_threshold_low = np.percentile(bootstrap_CCC_values, alpha/2*100)
CCC_threshold_high = np.percentile(bootstrap_CCC_values, (1-alpha/2)*100)

print(f"R² 95%置信区间: ({R_threshold_low}, {R_threshold_high})")

def calcuRatio(low,high,R_values):
    # 判断R0, R1, ..., R1000是否在置信区间内
    results = {}
    common_count = 0
    total_count = len(R_values)
    
    # for key, value in R_values.items():
    #     if low <= value <= high:
    #         results[key] = '常见情况'
    #         common_count += 1
    #     else:
    #         results[key] = '不常见情况'
    for key, value in R_values.items():
        if low <= value <= 1:
            results[key] = '常见情况'
            common_count += 1
        else:
            results[key] = '不常见情况'

    # 显示结果
    # for key, value in results.items():
    #     print(f"{key}: {value}")
    
    # 计算常见情况的比例
    
    common_ratio = common_count / total_count
    print(f"常见情况的比例: {common_count}/{total_count} = {common_ratio:.2f}")
    return common_ratio


test=calcuRatio(R_threshold_low,R_threshold_high, R_values)
test2=calcuRatio(CCC_threshold_low,CCC_threshold_high, CCC_values)
```
