```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate 5000 rows of data
rows = 5000

# First attribute: index
index = np.arange(1, rows + 1)

# Second attribute: sin(x) distribution with random noise
x_values = np.linspace(0, 10 * np.pi, rows)
sin_values = np.sin(x_values) + np.random.normal(0, 0.1, rows)

# Third attribute: 300 classes randomly assigned
classes = np.random.choice(range(1, 301), rows)

# Ensure each class has at least one row
for i in range(1, 301):
    if i not in classes:
        replace_index = np.random.randint(0, rows)
        classes[replace_index] = i

# Create the dataframe
df = pd.DataFrame({
    'Index': index,
    'SinValues': sin_values,
    'Class': classes
})

# Plot the scatter plot
plt.figure(figsize=(50, 6))
plt.scatter(df['Index'], df['SinValues'], alpha=0.5)
plt.title('Scatter Plot of Index vs. SinValues with Noise')
plt.xlabel('Index')
plt.ylabel('SinValues')
plt.grid(True)
plt.show()

df.head()

df.to_csv("./sin-out/generateData.csv")
```

```python
# Plot the frequency distribution of the third attribute (Class)
plt.figure(figsize=(50, 8))
plt.hist(df['Class'], bins=300, edgecolor='black', alpha=0.7)
plt.title('Frequency Distribution of Classes')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Introduce randomness and compute analysis
def random_model_1(sin_values):
    return sin_values + np.random.normal(0, 0.5, len(sin_values))

def random_model_2(sin_values):
    return sin_values * np.random.uniform(0.8, 1.2, len(sin_values))

def random_model_3(sin_values):
    return np.sin(sin_values * np.random.uniform(0.8, 1.2))

# Apply the random models
random_results_1 = random_model_1(df['SinValues'])
random_results_2 = random_model_2(df['SinValues'])
random_results_3 = random_model_3(df['SinValues'])


# Step 1Combine the results Random
combined_results = (random_results_1 + random_results_2 + random_results_3) / 3

# Step 2: Aggregate data based on the third attribute (Class)
df['RandomSinValues'] = combined_results
aggregated_data = df.groupby('Class').agg({'Index': 'mean', 'RandomSinValues': 'sum'}).reset_index()

# Step 3: Apply a nonlinear function (y = x^2) to the second column
aggregated_data['NonlinearValues'] = aggregated_data['RandomSinValues'] ** 2

# Print the resulting data
print(aggregated_data.head())

# Plot the scatter plot of the results
plt.figure(figsize=(10, 6))
plt.scatter(aggregated_data['Index'], aggregated_data['NonlinearValues'], alpha=0.5)
plt.title('Scatter Plot of Index vs. NonlinearValues')
plt.xlabel('Index')
plt.ylabel('NonlinearValues')
plt.grid(True)
plt.show()
aggregated_data.to_csv("./sin-out/generateDat_initialOut.csv")
```

```python
import warnings 

warnings.filterwarnings("ignore")

# Define the function to process data with randomness and apply nonlinear function
def process_data_with_randomness(data):
    def random_model_1(values):
        return values + np.random.normal(0, 0.5, len(values))

    def random_model_2(values):
        return values * np.random.normal(1, 0.1, len(values))

    def random_model_3(values):
        return values + np.random.uniform(-0.5, 0.5, len(values))

    # Apply all random models and average the results
    data['SinValuesRandom'] = (random_model_1(data['SinValues']) +
                               random_model_2(data['SinValues']) +
                               random_model_3(data['SinValues'])) / 3

    # Aggregate by the third column and sort by this attribute
    data_agg = data.groupby('Class').agg({'SinValuesRandom': 'sum'}).reset_index()
    data_agg = data_agg.sort_values(by='Class').reset_index(drop=True)

    # Apply a nonlinear function to the aggregated results
    data_agg['NonlinearValues'] = data_agg['SinValuesRandom'] ** 2

    return data_agg['NonlinearValues']

# Initialize the dataframe to store results
results_df = pd.DataFrame(index=range(0, 300))

# Run the process 1000 times and store the results
for i in range(0, 1001):
    processed_values = process_data_with_randomness(df.copy())
    results_df[f'Out{i}'] = processed_values

# Display the first few rows of the resulting dataframe
# results_df.head()
results_df.to_csv("./sin-out/outputs.csv")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# 
data_df = results_df

def lins_ccc(x, y):
    """Lin's Concordance Correlation Coefficient (CCC)"""
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    var_x = np.var(x)
    var_y = np.var(y)
    cov_xy = np.mean((x - mean_x) * (y - mean_y))
    rho = cov_xy / np.sqrt(var_x * var_y)
    ccc = (2 * rho * np.sqrt(var_x) * np.sqrt(var_y)) / (var_x + var_y + (mean_x - mean_y)**2)
    return ccc
    

# R0, R1, ..., R1000
R_values = {}
CCC_values={}
for i in range(1, 1001):
    R_values[f'R{i}'] = r2_score(data_df['Out0'], data_df[f'Out{i}'])
    CCC_values[f'R{i}'] = lins_ccc(data_df['Out0'], data_df[f'Out{i}'])

# bootstrapR
bootstrap_R_values = []
bootstrap_CCC_values=[]
num_bootstrap_samples = 10000

for _ in range(num_bootstrap_samples):
    sample_indices = np.random.choice(range(1, 1001), size=2, replace=True)
    # print(sample_indices)
    sample1 = data_df.iloc[:, sample_indices[0]]
    sample2 = data_df.iloc[:, sample_indices[1]]
    bootstrap_R_values.append(r2_score(sample1, sample2))
    bootstrap_CCC_values.append(lins_ccc(sample1, sample2))

# 
alpha = 0.05
bootstrap_R_values = np.sort(bootstrap_R_values)
R_threshold_low = np.percentile(bootstrap_R_values, alpha/2*100)
R_threshold_high = np.percentile(bootstrap_R_values, (1-alpha/2)*100)

bootstrap_CCC_values = np.sort(bootstrap_CCC_values)
CCC_threshold_low = np.percentile(bootstrap_CCC_values, alpha/2*100)
CCC_threshold_high = np.percentile(bootstrap_CCC_values, (1-alpha/2)*100)

print(f"R 95%: ({R_threshold_low}, {R_threshold_high})")

def calcuRatio(low,high,R_values):
    # R0, R1, ..., R1000
    results = {}
    common_count = 0
    total_count = len(R_values)
    
    for key, value in R_values.items():
        if low <= value <= high:
            results[key] = ''
            common_count += 1
        else:
            results[key] = ''

    # 
    # for key, value in results.items():
    #     print(f"{key}: {value}")
    
    # 
    
    common_ratio = common_count / total_count
    print(f": {common_count}/{total_count} = {common_ratio:.2f}")
    return common_ratio


test=calcuRatio(R_threshold_low,R_threshold_high, R_values)
test2=calcuRatio(CCC_threshold_low,CCC_threshold_high, CCC_values)
```
